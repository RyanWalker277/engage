```python
# Importing libraries
import numpy as np
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib
df1 = pd.read_csv("cars_engage_2022.csv")
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df2 = df1[['Engine_Location','Fuel_Tank_Capacity','Fuel_Type','Height','Length','Width','City_Mileage','Gears','Odometer','Speedometer','Seating_Capacity','Seats_Material','Audiosystem','Number_of_Airbags','Ex-Showroom_Price']]
df2['Fuel_Tank_Capacity'] = df2['Fuel_Tank_Capacity'].str.replace(' litres', '')
df2['Height'] = df2['Height'].str.replace(' mm', '')
df2['Length'] = df2['Length'].str.replace(' mm', '')
df2['Width'] = df2['Width'].str.replace(' mm', '')
df2['City_Mileage'] = df2['City_Mileage'].str.replace('?', '',regex=False)
df2['City_Mileage'] = df2['City_Mileage'].str.replace(' km/litre', '',regex=False)
df2['City_Mileage'] = df2['City_Mileage'].str.replace('12.5-', '',regex=False)
df2['City_Mileage'] = df2['City_Mileage'].str.replace(',', '.',regex=False)
df2['Ex-Showroom_Price'] = df2['Ex-Showroom_Price'].str.replace('Rs. ', '',regex=False)
df2['Ex-Showroom_Price'] = df2['Ex-Showroom_Price'].str.replace(',', '',regex=False)
df2['City_Mileage'].unique()
```

```python
df3 = df2
df3 = df3[df3.Number_of_Airbags != "Yes"]
df3 = df3[df3.Audiosystem != "Yes"]
df3 = df3[df3.Seating_Capacity != "Yes"]
df3 = df3[df3.Speedometer != "Yes"]
df3 = df3[df3.Odometer != "Yes"]
df3 = df3[df3.Gears != "7 Dual Clutch"]
df3 = df3[df3.Gears != "Single Speed Reduction Gear"]
```

```python
#dropping all nan values in engine location
df3 = df3[df3['Engine_Location'].notna()]
```

```python
#for fueltankcapacity

#filling all nan values with 0
df3['Fuel_Tank_Capacity'] = df3['Fuel_Tank_Capacity'].fillna(0)
#conveting fueltankcapacity to 0
df3['Fuel_Tank_Capacity'] = pd.to_numeric(df3['Fuel_Tank_Capacity'])
#calculating mean
mean_value = df3['Fuel_Tank_Capacity'].mean()
#filling mean
df3['Fuel_Tank_Capacity'].fillna(value = mean_value, inplace=True)
```

```python
#for numberofairbags

#filling all nan values with 0
df3['Number_of_Airbags'] = df3['Number_of_Airbags'].fillna(0)
#conveting fueltankcapacity to 0
df3['Number_of_Airbags'] = pd.to_numeric(df3['Number_of_Airbags'])
#calculating mean
mean_value_airbags = df3['Number_of_Airbags'].mean()
#filling mean
df3['Fuel_Tank_Capacity'].fillna(value=mean_value, inplace=True)
```

```python
#for seating capacity

#filling all nan values with 0
df3['Seating_Capacity'] = df3['Seating_Capacity'].fillna(0)
#conveting fueltankcapacity to 0
df3['Seating_Capacity'] = pd.to_numeric(df3['Seating_Capacity'])
#calculating mean
mean_value_seating = df3['Seating_Capacity'].mean()
#filling mean
df3['Seating_Capacity'].fillna(value=mean_value_seating, inplace=True)
```

```python
#for height

#filling all nan values with 0
df3['Height'] = df3['Height'].fillna(0)
#conveting fueltankcapacity to 0
df3['Height'] = pd.to_numeric(df3['Height'])
#calculating mean
mean_value_height = df3['Height'].mean()
#filling mean
df3['Height'].fillna(value=mean_value_height, inplace=True)
```

```python
#for width

#filling all nan values with 0
df3['Width'] = df3['Width'].fillna(0)
#conveting fueltankcapacity to 0
df3['Width'] = pd.to_numeric(df3['Width'])
#calculating mean
mean_value_width = df3['Width'].mean()
#filling mean
df3['Width'].fillna(value=mean_value_width, inplace=True)
```

```python
#for gears

#conveting fueltankcapacity to number
df3['Gears'] = pd.to_numeric(df3['Gears'])
#calculating mean
X = df3['Gears'].mean()
#filling mean
df3['Gears'].fillna(value= X , inplace=True)
```

```python
#for city milage

#conveting fueltankcapacity to number
df3['City_Mileage'] = df3['City_Mileage'].astype(float)
#calculating mean
X = df3['City_Mileage'].mean()
#filling mean
df3['City_Mileage'].fillna(value= X , inplace=True)

#dropping the row containing a logically wrong value

df3.drop(df3.index[df3['City_Mileage'] == 1.845000 ], inplace = True)
```

```python
#filling all values in speedometer and odometer with analog 
df3[['Odometer' , 'Speedometer']] = df3[['Odometer' , 'Speedometer']].fillna("Analog")
```

```python
df3.dropna( subset=['Seats_Material'] , inplace= True)
df3.dropna( subset=['Audiosystem'] , inplace= True)
```

```python
#hot encoding part
Engine_Location = pd.get_dummies(df3.Engine_Location)
Fuel_Type = pd.get_dummies(df3.Fuel_Type)
Odometer = pd.get_dummies(df3.Odometer)
Speedometer = pd.get_dummies(df3.Speedometer)
Seats_Material = pd.get_dummies(df3.Seats_Material)
Audiosystem = pd.get_dummies(df3.Audiosystem)
```

```python
df4 = pd.concat([df3,Engine_Location,Fuel_Type,Odometer,Speedometer,Seats_Material,Audiosystem],axis='columns')
# df4 = pd.concat([df3,Engine_Location.drop('Rear, Transverse',axis='columns'),Fuel_Type.drop('Petrol',axis='columns'),Odometer.drop('Digital, Analog',axis='columns'),Speedometer.drop('Digital, Analog',axis='columns'),Seats_Material.drop('Vinyl',axis='columns'),Audiosystem.drop('USB & Aux-in',axis='columns')],axis='columns')
# renaming the ex-showroom price coloumn
df4.rename(columns = {'Ex-Showroom_Price':'price'}, inplace = True)
#df3.shape
#df3['Gears'].unique()

# #print(df3['Height'])
#print(df3['Seats_Material'].value_counts())
#df3.isna().sum()
# df4
```

```python
# dropping the hard encoded coloumns
df4.drop('Engine_Location',axis='columns' , inplace = True)
df4 = df4.drop('Fuel_Type',axis='columns')
df4 = df4.drop('Odometer',axis='columns')
df4 = df4.drop('Speedometer',axis='columns')
df4 = df4.drop('Seats_Material',axis='columns')
df4 = df4.drop('Audiosystem',axis='columns')
df4
```

```python
X = df4.drop('price' , axis = 'columns')
y = df4.price
```

```python
# implemnting decision tree
from sklearn import tree
model = tree.DecisionTreeClassifier(splitter='best')
model.fit(X.values,y) #training
```

```python
def predict_price(Fuel_Tank_Capacity,Height,Length,Width,City_Mileage,Gears,Seating_Capacity,Number_of_Airbags,Engine_Location,Fuel_Type,Odometer,Speedometer,Seats_Material,Audiosystem):    
    engine_index = np.where(X.columns==Engine_Location)[0][0]
    fuel_index = np.where(X.columns==Fuel_Type)[0][0]
    odometer_index = np.where(X.columns==Odometer)[0][0]
    speedometer_index = np.where(X.columns==Speedometer)[0][0]
    seats_index = np.where(X.columns==Seats_Material)[0][0]
    audiosystem_index = np.where(X.columns==Audiosystem)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = Fuel_Tank_Capacity
    x[1] = Height
    x[2] = Length
    x[3] = Width
    x[4] = City_Mileage
    x[5] = Gears
    x[6] = Seating_Capacity
    x[7] = Number_of_Airbags
    
    if engine_index >= 0:
        x[engine_index] = 1
    if fuel_index >= 0:
        x[fuel_index] = 1
    if odometer_index >= 0:
        x[odometer_index] = 1
    if speedometer_index >= 0:
        x[speedometer_index] = 1
    if seats_index >= 0:
        x[seats_index] = 1
    if audiosystem_index >= 0:
        x[audiosystem_index] = 1

    return model.predict([x])[0]
```
